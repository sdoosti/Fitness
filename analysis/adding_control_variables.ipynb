```python
import pandas as pd
import numpy as np
import os
```

```python
Path = "C:/Users/doosti/OneDrive - Chapman University/Research/Research Projects/Fitness/Data/"
print(Path)
```

```python
# load data
channels = pd.read_csv(Path+'channel_trends_to_stata_performance_v3.csv')
print(channels.head())
```

```python
# loading raw data
videos = pd.read_csv(os.path.join(Path, "videos_101223.csv"))
print(videos.head())
```

```python
# creating number of uploads for the past 7 and 30 days
# number of uploads (rows) in the past 7 days from videos dataframe
channels['uploads_7d'] = channels.groupby('creator_id').uploads.rolling(window=7, min_periods=1).sum().reset_index(0, drop=True)
# number of uploads (rows) in the past 30 days from videos dataframe
channels['uploads_30d'] = channels.groupby('creator_id').uploads.rolling(window=30, min_periods=1).sum().reset_index(0, drop=True)
print(channels[['uploads', 'uploads_7d', 'uploads_30d']].describe())
print(channels[['creator_id', 'uploads', 'uploads_7d', 'uploads_30d']].head(30))
```

```python
channels.columns
```

```python
# number of uploads in the past 7 days by challenge: separate number of uploads for chanllege videos and non-challenge videos
channels['uploads_7d_challenge'] = channels.groupby('creator_id').challenge_dummy.rolling(window=7, min_periods=1).sum().reset_index(0, drop=True)
# number of uploads (rows) in the past 30 days from videos dataframe
channels['uploads_30d_challenge'] = channels.groupby('creator_id').challenge_dummy.rolling(window=30, min_periods=1).sum().reset_index(0, drop=True)
# non-challenge videos
channels['uploads_7d_nonchallenge'] = channels['uploads_7d'] - channels['uploads_7d_challenge']
channels['uploads_30d_nonchallenge'] = channels['uploads_30d'] - channels['uploads_30d_challenge']

print(channels[['uploads_7d_challenge', 'uploads_30d_challenge', 'uploads_7d_nonchallenge', 'uploads_30d_nonchallenge']].describe())
```

```python
# creating total number of minutes posted for the last 7 and 30 days
# preparing duration column
videos['date'] = videos['publish_date']
grouped = videos.groupby(['creator_id', 'date']).agg({'duration': 'sum'}).reset_index()
channels.rename(columns={'duration': 'challenge_duration'}, inplace=True)
channels['total_duration'] = channels.merge(grouped, on=['creator_id', 'date'], how='left').duration.fillna(0)
# total_duration for the past 7 days
channels['total_duration_7d'] = channels.groupby('creator_id').total_duration.rolling(window=7, min_periods=1).sum().reset_index(0, drop=True)
# total_duration for the past 30 days
channels['total_duration_30d'] = channels.groupby('creator_id').total_duration.rolling(window=30, min_periods=1).sum().reset_index(0, drop=True)
channels.describe()
```
