```python
comments_day = pd.read_csv(Path + "channel_day_scores_010324.csv")
print(comments_day.shape)
print(comments_day.head())
```

```python
# loading the raw comment data
comments = pd.read_csv(Path + "labeled_comments_scores_030124.csv")
print(comments.shape)
print(comments.head())
```

```python
# loading videos for creator id
videos = pd.read_csv(os.path.join(Path, "videos_101223.csv"))
```

```python
videos.head()
```

```python
# TEMP
# merging the challenge information to test hypotheses
comments_stats = comments.merge(videos, on='video_id', how='left')
grouped_commets = comments_stats.groupby('video_id').agg({'user_id':'count','challenge':'first'})
print(grouped_commets.head())
print(grouped_commets.groupby('challenge').agg({'user_id':'mean'}))
```

```python
comments['creator_id'] = comments.merge(videos,how='left',on='video_id').creator_id.values
comments = comments.dropna(subset=['creator_id'])
comments.isnull().sum()
```

```python
grouped = comments.groupby(['creator_id','date'])[['no_likes','no_replies']].sum().reset_index()
print(grouped.head())
comments_day = comments_day.merge(grouped,how='left',on=['creator_id','date'])
print(comments_day.head())
```

```python
# adding likes and replies for the competence
grouped = comments[comments.label=="Competence"].groupby(['creator_id','date'])[['no_likes','no_replies']].sum().reset_index()
# changing the column names
grouped.columns = ['creator_id','date','competence_likes','competence_replies']
comments_day = comments_day.merge(grouped,how='left',on=['creator_id','date'])
# adding likes and replies for the social
grouped = comments[comments.label=="Social"].groupby(['creator_id','date'])[['no_likes','no_replies']].sum().reset_index()
# changing the column names
grouped.columns = ['creator_id','date','social_likes','social_replies']
comments_day = comments_day.merge(grouped,how='left',on=['creator_id','date'])
print(comments_day.head())
```
